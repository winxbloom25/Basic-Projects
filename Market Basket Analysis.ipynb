```python
List_item = []
for i in range(row):
    temp = []
    for j in item_list:
         if new_data.loc[i,j]>=1:
                temp.append(j)
    List_item.append(temp)
    
List_item    
```

```python
new_data1['Item_List'] = List_item
new_data1.head()
```

```python
basket_data =new_data1[['Item_List']]
basket_data.head()
```

```python
trans = new_data1.drop(['Item_List'], axis=1)
trans.head()
```

```python
def encode_units(x):
    if x <1:
        return 0
    if x >= 1:
        return 1
    
basket_trans = trans.applymap(encode_units)
basket_trans.head()
```

### Association Rule Mining

```python
whole_milk_data = basket_trans.loc[basket_trans['whole milk']==1]
whole_milk_data
```

```python
frequent_items = apriori(whole_milk_data, min_support = 0.15, use_colnames=True)
frequent_items
```

```python
whole_milk_rules = association_rules(frequent_items, metric = 'lift', min_threshold=1)
```

```python
whole_milk_rules.sort_values(['lift', 'support'],ascending=False).reset_index(drop=True)
```

```python
def frequently_bought_together(item):
    item_df = basket_trans.loc[basket_trans[item]==1]
    frequent_itemsets = apriori(item_df, min_support=0.15, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric = 'lift', min_threshold=1)
    rules.sort_values(['lift', 'support'], ascending=False).reset_index(drop=True)
    print('Items frequently bought together with {0}'.format(item))
    return rules['consequents'].unique()[:6]
```

```python
frequently_bought_together('salty snack')
```
